```python
import pandas as pd
grades = pd.read_csv('grades.csv')
grades
```

```python
df = pd.DataFrame(grades)
df['year'] =  pd.DatetimeIndex(df['submit_time']).year
df['date'] =  pd.DatetimeIndex(df['submit_time']).date
df['day'] =  pd.DatetimeIndex(df['submit_time']).day
df['month'] =  pd.DatetimeIndex(df['submit_time']).month
df['hour'] =  pd.DatetimeIndex(df['submit_time']).hour
df['minute'] =  pd.DatetimeIndex(df['submit_time']).minute
df
```

```python
df['FileType'] = df.submission.str.rsplit('.', 1).str[-1]
```

```python
df
```

```python
df.count()
```

```python
mask_by_file_type_zip = df.loc[df["FileType"] == "zip", :]
mask_by_file_type_zip
```

```python
mask_by_file_type_zip.count()
```

```python
mask_by_file_type_zip["submission"].count()/df["submission"].count()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
mask_by_first_deadline = df.loc[(df['date'] > '2017-01-03'),:]
mask_by_first_deadline
```

```python
mask_by_first_deadline["submission"].count()/df["submission"].count()
```

```python
grouped_by_date = mask_by_first_deadline.groupby('date')
count_by_date = grouped_by_date['date'].count()
print(count_by_date)
```
