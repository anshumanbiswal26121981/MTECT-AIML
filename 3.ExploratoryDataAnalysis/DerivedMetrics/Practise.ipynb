```python
# Derived Variable
# Description
# The given dataset 'cust_rating' has 3 columns i.e 'rating', ' food_rating', 'service_rating'. Create a new variable 'avg_rating'.


import pandas as pd
cust_rating = pd.read_csv('https://media-doselect.s3.amazonaws.com/generic/GvOkzy5MLdy9zJ8vXMArwpRvN/rating_final.csv')

cust_rating['avg_rating'] = round(cust_rating.loc[:,[c for c in cust_rating.columns if "rating" in c]].mean(axis=1))

print(cust_rating.head(10))
```

```python
odi = pd.read_csv("odi-batting.csv")
odi
```

```python
df = pd.DataFrame(odi)
df
```

```python
df.columns
```

```python
mask = df['Runs'] >= 100
filtered_data = df.where(mask)
```

```python
grouped_data = filtered_data.groupby('Player')
```

```python
centuries_by_player = grouped_data['Runs'].count()
print(centuries_by_player)
```

```python

centuries_by_player.nlargest(3) 
```

```python
df["StrikeRate"] = df["Runs"] / df["Balls"]
df
```

```python
mask_by_century = df['Runs'] >= 100
filtered_data_by_century = df.where(mask)
```

```python
filtered_data_by_century
```

```python
grouped_data_by_century = filtered_data_by_century.groupby('Player')
```

```python
high_strike_rate_player_by_centuries = grouped_data_by_century['StrikeRate'].sum()
print(high_strike_rate_player_by_centuries)
```

```python
high_strike_rate_player_by_centuries.nlargest(3)
```

```python
df['Year'] =  pd.DatetimeIndex(df['MatchDate']).year
df
```

```python
mask_by_country = df.loc[df["Country"] == "India", :]
mask_by_country
```

```python
mask_india_player_century = mask_by_country.loc[mask_by_country["Runs"] >= 100, :]
mask_india_player_century
```

```python
max_runs_scored_in_year = grouped_data_by_year['Runs'].sum()
print(max_runs_scored_in_year)
```
