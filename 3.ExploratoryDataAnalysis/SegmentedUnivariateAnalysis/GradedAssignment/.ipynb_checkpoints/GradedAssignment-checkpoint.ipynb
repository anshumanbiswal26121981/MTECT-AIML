```python
import pandas as pd
nsa = pd.read_csv('EDA_census.csv')
nsa
```

```python
df = pd.DataFrame(nsa)
df
```

```python
Total_illiterate_female = df[df['Age-group']=='20-24'].Illiterate_Females.sum()
Total_illiterate_female
```

```python
Total_females = df[df['Age-group']=='20-24'].Total_Females.sum()
Total_females
```

```python
#percentage
Total_illiterate_female/Total_females
```

```python
df1 = df.groupby(['Age-group'])['Total_Persons'].sum()
df1
```

```python
df['female_Literacy_rate']= df['Literate_Females']/df['Literate_Persons']
df
```

```python
df2 = df.groupby(['Area Name'])['female_Literacy_rate'].sum()
df2
```
